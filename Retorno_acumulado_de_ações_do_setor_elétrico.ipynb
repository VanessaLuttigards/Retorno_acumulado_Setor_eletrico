```python
acoes_eletro_filter = ['ALUP11.SA',
'AURE3.SA',
'CMIG4.SA',
'COCE5.SA',
'CPFE3.SA',
'CPLE6.SA',
'ELET3.SA',
'ENEV3.SA',
'ENGI11.SA',
'EGIE3.SA',
'EQTL3.SA',
'NEOE3.SA',
'LIGT3.SA',
]
```

## Extraindo dados

```python
setor_eletrico = yf.download(acoes_eletro_filter, start = '2023-01-01')['Adj Close']
```

```python
setor_eletrico.index = pd.to_datetime(setor_eletrico.index)
```

```python
round(setor_eletrico,2)
```

```python
setor_eletrico.plot(figsize=(8,6))
```

## Fazendo a normalização

```python
setor_eletrico_norm = setor_eletrico/setor_eletrico.iloc[0]
```

```python
setor_eletrico_norm.plot(figsize=(10, 6) )
plt.legend(loc='upper left')
plt.show()
```

##  Pequena analise da empresa de melhor performance

```python
cemig = yf.download('CMIG4.SA', start = '2024-01-01')
```

###Vamos plotar o preço e o volume desde o início de 2024

```python
fig = plt.figure(figsize = (10,5))
gs = fig.add_gridspec(nrows = 6, ncols = 1)

ax1 = fig.add_subplot(gs[0:5, 0])
ax2 = fig.add_subplot(gs[5, 0])

ax1.plot(cemig.Close)
ax2.bar(cemig.index, cemig.Volume)

ax1.set_title('Fechamento | Volume | Cotação de CMIG4')
```
